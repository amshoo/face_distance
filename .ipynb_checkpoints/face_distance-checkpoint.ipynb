```python
import tkinter as tk
from tkinter import *
from tkinter import font

#for python 2: import Tkinter
from PIL import ImageTk, Image
from tkinter import filedialog
import os

root = tk.Tk()
root.resizable(0, 0) 

HEIGHT=510
WIDTH=900


def openfn():
    filename = filedialog.askopenfilename(title='open')
    return filename

r_image=[]
def r_open_img():
    x = openfn()
    img = Image.open(x)
    img = img.resize((200, 200), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    panel = Label(frame3, image=img)
    panel.image = img
    panel.pack()
    
    r_image.append(x)

l_image=[]
def l_open_img():
    y = openfn()
    img = Image.open(y)
    img = img.resize((200, 200), Image.ANTIALIAS)
    img = ImageTk.PhotoImage(img)
    panel = Label(frame2, image=img)
    panel.image = img
    panel.pack()
    
    l_image.append(y)
      
import face_recognition as fr

def face_mach(left_image, right_image):
    #load known image and encode face
    left_image = fr.load_image_file(left_image)
    left_face_encoding = fr.face_encodings(left_image)[0]

    #load unknown image and encode face
    right_image = fr.load_image_file(right_image)
    right_face_encoding = fr.face_encodings(right_image)[0]
    #./img/unknown/d-trump.jpg
    known_encodings = [
    right_face_encoding
    ]

    # Compare faces
    result = fr.compare_faces([left_face_encoding], right_face_encoding)
    
    face_distances = fr.face_distance(left_face_encoding, known_encodings)


    if result[0]:
        rst = 'Same Person with {}% distance'.format(face_distances*100)
    else:
        rst = 'NOT Same Person with {}% distance'.format(face_distances*100)        
    return rst
        

def images():
    l_x = l_image 
    l_x = ''.join(map(str, l_x))
    
    r_x = r_image 
    r_x = ''.join(map(str, r_x))
    
    z = face_mach(l_x,r_x)
    
    
   
    print_result.configure(text=face_mach(l_x,r_x))
    #print(face_mach(l_x,r_x))
    
    
#canvas------------------------------------------------------------------------
canvas =tk.Canvas(root,height=HEIGHT, width=WIDTH)
canvas.pack()

title = Label(root,text="Face_Match", font=("Helvetica", 20))
title.place(x=300, y=10)

#frame 1---------------------------------------------------------
frame = tk.Frame(root, highlightbackground="gray", 
                 highlightcolor="gray", 
                 highlightthickness=1.5, 
                 bd=10)

frame.place(x=400, y=60 ,
            width=700, height=300, 
            anchor='n')

#frame 2-----------------------------------------------------------
frame2 = tk.Frame(frame, highlightbackground="gray", 
                  highlightcolor="white", 
                  highlightthickness=1.5, 
                  bd=10)
frame2.place(x=100,y=50, width=200, 
             height=200, 
             anchor='n')
#frame 3-----------------------------------------------------------
frame3 = tk.Frame(frame, highlightbackground="gray", 
                  highlightcolor="white", 
                  highlightthickness=1.5, 
                  bd=10)
frame3.place(x=520,y=50, width=200, 
             height=200, 
             anchor='n')
#frame 4-----------------------------------------------------------
frame4 = tk.Frame(root, highlightbackground="gray", 
                  highlightcolor="white", 
                  highlightthickness=1.5, 
                  bd=10)
frame4.place(x=400,y=370, width=700, 
             height=100, 
             anchor='n')
#btn1 ----------------------------------------------------------------------
btn = Button(frame, text='Load left image', command=l_open_img).place(x=100, anchor='n')
btn = Button(frame, text='Load right image', command=r_open_img).place(x=520, anchor='n')
match_btn = Button(frame, text='match image', command=images).pack()


#print result
print_result = Label(frame4, text='', font=("Helvetica", 10)) 
print_result.place(x=20,y=20)

root.mainloop()
```
