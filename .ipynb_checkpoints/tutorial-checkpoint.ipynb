```python
import tkinter as tk
from tkinter import *
from tkinter.ttk import *
from tkinter import font
from tkinter import messagebox

#for python 2: import Tkinter

root = tk.Tk()
root.resizable(0, 0) 

HEIGHT=510
WIDTH=1150

#Definetions
def selected_data_list():

    selected_data_list=[]
    
    symtoms = ( gender.get(),
                famliy.get(),
                diabetes.get(),
                symtom1.get(),
                symtom2.get(),
                symtom3.get(),
                symtom4.get(),
                symtom5.get(),
                symtom6.get(),
                symtom7.get(),
                symtom8.get())
    for m in  symtoms:
        selected_data_list.append(m)
        
    return selected_data_list

#conditon to Update Selections--------------------------------------------------------------------
def update():
    
    personal_info_list=[]
    
    personal_data = ("Name: "+ name.get(),
                     "Address: "+ address.get(),
                     "Phone: "+ phone.get(),
                     "Age: "+ age.get())
    
    symtoms = (selected_data_list())#get the selected symtoms and state.
  
    personal_info = personal_data
    selected_data = symtoms
    
    for n in personal_info:
        personal_info_list.append(n)
        
        
#if the personal data entery are empty return Error MessegeBox-------------------------------------------        
    if len(name.get()) == 0 or len(address.get()) == 0 or len(phone.get()) == 0:  
        messagebox.showinfo("Missing Information", "Pls fill up your personal data")        
        
    else:     
        df = pd.DataFrame(selected_data, columns=[X],index=['Gender','Family','diabetis','symtom1', 'symtom2','symtom3', 'symtom4','symtom5', 'symtom6','symtom7', 'symtom8'])
    
        print_personal_data.configure(text=personal_info_list)
       # print_selected_data.configure(text=df)
        result.delete(1.0, tk.END)
        result.insert(tk.END, str(df))

#conditon to start processing Frames------------------------------------------------------------------
def process():   
    if len(name.get()) == 0 or len(address.get()) == 0 or len(phone.get()) == 0:  
        messagebox.showinfo("Missing Information", "Pls fill up your personal data")        
        
    else:
        result=messagebox.askquestion("Process", "Are You Sure you want to continuoue", icon='warning')
        if result == 'yes':
            update_symtoms.config(state="disabled")
            process_btn.config(state="disabled")
            diagnose_btn.config(state="normal")
        else:
            btn.config(state="normal")

#Diagnose ------------------------------------------------------------------------------------------------
import pandas as pd
def diagnose():    
    #Fuzzy RULE
    data = selected_data_list()
    df = pd.DataFrame(data, index=['Gender','Family','diabetis','symtom1', 'symtom2','symtom3', 'symtom4','symtom5', 'symtom6','symtom7', 'symtom8'])
    print(df)
       
    
#canvas------------------------------------------------------------------------
canvas =tk.Canvas(root,height=HEIGHT, width=WIDTH)
canvas.pack()

title = Label(root,text="Diagnosis Expert System", font=("Helvetica", 20))
title.place(x=300, y=10)

#frame 1---------------------------------------------------------
frame = tk.Frame(root, highlightbackground="gray", 
                 highlightcolor="gray", 
                 highlightthickness=1.5, 
                 bd=10)

frame.place(x=400, y=60 ,
            width=700, height=300, 
            anchor='n')

#frame 2-----------------------------------------------------------
frame2 = tk.Frame(frame, highlightbackground="gray", 
                  highlightcolor="white", 
                  highlightthickness=1.5, 
                  bd=10)
frame2.place(relx=0.5, relwidth=1, 
             height=40, 
             anchor='n')

#Select Patient Age ----------------------------------------------------------------------
age = Combobox(frame2)
age['values']= ('Age Range','0-20', '21-40', '41-60', '61-80', '81-above') 
age.current(0) 
age.grid(column=1, row=0,padx=10)

#Select Patient Gender-------------------------------------------------------------------
gender = Combobox(frame2)
gender['values']= ('Gender','Male', 'Female') 
gender.current(0)
gender.grid(column=3, row=0,padx=10)

#Select yes if Patient Famliy History has someone with Glaucoma --------------------------
famliy = Combobox(frame2)
famliy['values']= ('Famliy History','Yes', 'No') 
famliy.current(0) 
famliy.grid(column=5, row=0,padx=10)

#Select yes, if Patient has diabetes---------------------------------------------
diabetes = Combobox(frame2)
diabetes['values']= ('Diabetes','Yes', 'No') 
diabetes.current(0) 
diabetes.grid(column=7, row=0,padx=10)

#frame 3------------------------------------------------------------
def Q_frame1(que,op1,op2,op3,op4,left,top):  
    #Frame  
    Q_frame1 = tk.Frame(frame,bd=10)
    
    Q_frame1.place(x=left, y=top,  
                   width=150, height=110, 
                   anchor='n')
    #Label------------------------------------------------------------------------
    label=Label(Q_frame1, text=que)
    label.pack()
    
    #symtoms_state----------------------------------------------------------------------
    symtoms_state = Combobox(Q_frame1)
    symtoms_state['values']= (op1,op2,op3,op4) 
    symtoms_state.current(0) #set the selected item 
    symtoms_state.pack()
    
    return symtoms_state


#2 options    
def Q_frame2(que,op1,op2,left,top):  
    #Frame  ---------------------------------------------------------------------------
    Q_frame2 = tk.Frame(frame,bd=10)
    Q_frame2.place(x=left, y=top,  
                   width=150, height=110, 
                   anchor='n')
    #Label------------------------------------------------------------------------------
    label=Label(Q_frame2, text=que)
    label.pack()
    
    #symtoms state Frame-----------------------------------------------------------------
    symtoms_state = Combobox(Q_frame2)
    symtoms_state['values']= (op1,op2) 
    symtoms_state.current(0) #set the selected item 
    symtoms_state.pack()

    return symtoms_state 

top1 = 60
top2 = 170


#First line-----------------------------------------------------------------------------------
left=100
symtom1 = Q_frame1('Red EYE','Mild', 'Moderate', 'Severe', 'Very severe',left,top1)
left= left+ 160
symtom2 = Q_frame1('EYE Pain','Mild', 'Moderate', 'Severe', 'Very severe',left,top1)
left= left+ 160
symtom3 = Q_frame2('Lost of Visual Acuty','Yes','No',left,top1)
left= left+ 160
symtom4= Q_frame2('Headache','Yes','No',left,top1)


#second line-------------------------------------------------------------------------------------
left=100
symtom5=Q_frame1('Halos around lights','Mild', 'Moderate', 'Severe', 'Very severe',left,top2)
left=left+160
symtom6=Q_frame2('Nausea and Vomiting','Yes','No',left,top2)
left=left+160
symtom7=Q_frame2('Abdominal Pain','Yes','No',left,top2)
left=left+160
symtom8=Q_frame1('Lost of Vision','Mild','Mild','Mild','Mild',left,top2)

#frame 4---------------------------------------------------------------
frame4 = tk.Frame(root,  highlightbackground="gray", 
                  highlightcolor="white", 
                  highlightthickness=1.5, bd=10)

frame4.place(x=860, y=60 ,
             width=200, height=220, 
             anchor='n')

#create entry fields----------------------------------------------------
def personal_info(text, lrow,erow):    
    L1 = Label(frame4,text=text)
    L1.grid(column=0, row=lrow)
    E1 = Entry(frame4, width=28)
    E1.grid(column=0, row=erow)
    return E1

#Call personal_info(text, lrow,erow)------------------------------------
name = personal_info("Patient Name",0,1)
address= personal_info('Patient Address',2,3)
phone =personal_info('Patient Phone Number',4,5)

#frame 5--------------------------------------------------------------
frame5 = tk.Frame(frame4, highlightbackground="gray", 
                 highlightcolor="white", 
                 highlightthickness=1.5,
                 bg='white', bd=10)
frame5.place(y=150,relx=0.5, 
            relwidth=1, height=50, 
            anchor='n')

#update Entry Button----------------------------------------------------
update_symtoms = Button(frame5, text="Update Symtoms", command=update)
update_symtoms.place(relx=0.5,relwidth=1, height=30, anchor='n')


#frame 6---------------------------------------------------------------
frame6 = tk.Frame(root,highlightbackground="gray", 
                 highlightcolor="white", 
                 highlightthickness=1.5,
                  bd=10)
frame6.place(x=400, y=365 , width=700, height=100, anchor='n')

label = Label(frame6,text="Diagnois Result: ")
label.grid(column=0, row=1)

#Label print patient personal information------t = Text(parent, width=40, height=10)----------------------------
print_personal_data = Label(frame6, text="") 
print_personal_data.grid(column=0, row=2)

#Label print symptoms and selected symtom state ---------------------------- 
print_selected_data = Label(frame6, text="") 
print_selected_data.grid(column=0, row=4)

#frame 7----------------------------------------------------------------

frame7 = tk.Frame(root, highlightbackground="gray", 
                 highlightcolor="white", 
                 highlightthickness=1.5,
                 bd=10)
frame7.place(x=960, y=290 ,
             width=200, height=75,
             anchor='n')

#Analyse/Diagnose Button -------------------------------------------------
diagnose_btn = Button(frame7, text="Analyse/Diagnose",command=diagnose)
diagnose_btn.place(relx=0.5, y=10 , width=160, height=30, anchor='n')
diagnose_btn.config(state="disabled")


#Frame8--------------------------------------------------------------------------
frame8 = tk.Frame(root, highlightbackground="gray", 
                 highlightcolor="white", 
                 highlightthickness=1.5,
                 bd=10)
frame8.place(x=1045, y=60 ,
             width=150, height=220, 
             anchor='n')

result = Text(frame8, width=40, height=10, font=("Helvetica", 8))
result.pack()


#process Button --------------------------------------------------------
process_btn = Button(frame8, text="Process Result",command=process)
process_btn.place(relx=0.5,y=160, relwidth=1, height=30, anchor='n')



root.mainloop()
```
